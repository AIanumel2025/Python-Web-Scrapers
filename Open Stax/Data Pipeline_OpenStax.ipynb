```python
# Locate question blocks and extract text
def open_stax_scraper():
    algebra=requests.get('https://openstax.org/books/algebra-1/pages/1-5-6-practice')
    algebra1=BeautifulSoup(algebra.text)
    alg_text=algebra1.find_all('div', class_='os-embed autogenerated-content')
    
    def question_text(t):
        statement = t.find('li')
        if statement:
            return statement.get_text(separator=" ", strip=True)
        return ""
    
    def answer_choices_text(t):
        choices = t.find_all('li', class_='os-choice')
        return [c.get_text(strip=True) for c in choices]
    
    # extract images
    def images(t):
        imgs=t.find_all('img')
        return [img["src"] for img in imgs if img.get("src")]
    
    questions = []
    for idx, t in enumerate(alg_text, start=1):
        record = {
        "Question_number": idx,
        "Question_text": question_text(t),
        "Possible_choices": answer_choices_text(t),
        "Image_urls": images(t),
        "Has_image": len(images(t)) > 0,
        "Source": "OpenStax"}
        
        questions.append(record)
    
    # convert to dataframe
    return pd.DataFrame(questions)

# call function
df=open_stax_scraper()
df
```
