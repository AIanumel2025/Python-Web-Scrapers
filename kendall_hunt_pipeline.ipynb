```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import uuid
```

```python
requests.get('https://im.kendallhunt.com/HS/students/1/2/2/practice.html')
```

```python
def generate_id():
    return str(uuid.uuid4())
    
def kendall_hunt_scraper():
    ken = requests.get('https://im.kendallhunt.com/HS/students/1/2/2/practice.html')
    kendall = BeautifulSoup(ken.text, "html.parser")
    kendall_hunt = kendall.find('div', class_='im-c-container im-c-container--content im-c-container--padded')
    question_divs = kendall_hunt.find_all('div', class_='im-c-row')

    def images(q):
        imgs = q.find_all('img')
        return [img["src"] for img in imgs if img.get("src")]

    questions2 = []
    for q in question_divs:
        num = q.find('h3', class_='im-c-heading--xsb')
        text = q.find('div', class_='im-c-row__body')
        
        record = {
            "ID": generate_id(),
            "Question_number": num.get_text(strip=True) if num else None,
            "Question_text": text.get_text(strip=True) if text else None,
            "Image_urls": images(q),
            "Has_image": len(images(q)) > 0,
            "Source": "Kendall Hunt"
    }
        questions2.append(record)
        
    return pd.DataFrame(questions2)

    
df=kendall_hunt_scraper()
df
```
